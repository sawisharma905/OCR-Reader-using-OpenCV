This notebook has the code written for making an Optical Character Reader using python and its libraries. The given code enables the user to take an image as an input and read the characters from the given image and show it up to the user. 
The model makes use of the major python libraries such as OpenCV, Google tesseract, and matplotlib.

```python
#importing OpenCV library
import cv2
```

```python
#Google tesseract is an OCR engine - this is available open source, free of cost
import pytesseract
```

```python
#importing matplotlib so that when needed, I can print the required images import the images inline
import matplotlib.pyplot as plt
```

```python
#reading the file whose data is to be read
data1=cv2.imread("task_img1.jpg")
data2=cv2.imread("task_img2.jpg")
print("Data file imported !!")
```

```python
#showing up first image of the analysis
plt.imshow(data1)
plt.title("Actual first image")
plt.show()
#showing up the second image of analysis
plt.imshow(data2)
plt.title("Actual second image")
plt.show()
```

### ANALYSIS 1: Reading the data from the real image (without any enhancements)

```python
#converting the original image to the text data
textdata1=pytesseract.image_to_string(data1)
textdata2=pytesseract.image_to_string(data2)
print(textdata1, textdata2)
# print(textdata2)
```

### ANALYSIS 2: Reading the image after converting into grayscale 

```python
#converting the image to greyscale for better quality
graydata1=cv2.cvtColor(data1 , cv2.COLOR_RGB2GRAY)
graydata2=cv2.cvtColor(data2 , cv2.COLOR_RGB2GRAY)

plt.imshow(graydata1)
plt.title("First grayscale image")
plt.show()

plt.imshow(graydata2)
plt.title("Second grayscale image")
plt.show()
```

```python
#adjusting the brightness of the image
#cv2.threshold, cv2.adaptiveThreshold : If pixel value is greater than a threshold value, it is assigned one value (may be white), else it is assigned another value (may be black).
#The image may has different lighting conditions in different areas. In that case, we go for adaptive thresholding. 

bright1= cv2.adaptiveThreshold(graydata1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C , cv2.THRESH_BINARY, 21,11)
bright2= cv2.adaptiveThreshold(graydata2, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C , cv2.THRESH_BINARY, 21,11)


plt.imshow(bright1)
plt.title("Brightened first grayscale image")
plt.show()

plt.imshow(bright2)
plt.title("Brightened second grayscale image")
plt.show()
```

```python
#converting the image to the text data
textdata3=pytesseract.image_to_string(bright1)
textdata4=pytesseract.image_to_string(bright2)
```

```python
print(textdata3 , textdata4)
```
